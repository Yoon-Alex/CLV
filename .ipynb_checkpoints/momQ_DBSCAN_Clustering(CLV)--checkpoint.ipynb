```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyodbc
```

```python
from sklearn.cluster import DBSCAN
import seaborn as sns
```

```python
dataset = pd.read_csv("CLV.csv",engine = 'python')
```

```python
dataset.shape
```

```python
df = dataset.sample(n=50000, random_state=1)
```

```python
df.replace({"GD": {"NO_GD": 0, "남자": 1, "여자": 2 }}, inplace=True)
```

```python
df.columns
```

```python
# 열별로 scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
a = ['score_predicted', 'revenue_predicted', 'prob_predicted', 'GD', 'KID_YN'] # 'GD', 'KID_YN'
```

```python
X = df_scaled[a].values
```

```python
clustering = DBSCAN(eps=0.4, min_samples= 110) #110
```

```python
y_hc = clustering.fit_predict(X)
```

```python
df['result']= y_hc
```

```python
df.groupby('result').mean().loc[:,a].sort_values(['score_predicted'], ascending = False)
```

```python
df.groupby('result').mean().loc[:,a+['AMT', 'TR', 'ATV', 'KIDS', 'JDT', 'RECENCY', 'MAX_AMT']].sort_values(['score_predicted', 'TR', 'RECENCY'], ascending = False)
```
