```python
df = dataset
```

```python
df.replace({"GD": {"NO_GD": 0, "남자": 1, "여자": 2 }}, inplace=True)
```

```python
# 열별로 scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
X = df_scaled[a]
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X.values)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Applying K-Means to dataset
kmeans = KMeans(n_clusters=4,init='k-means++',max_iter=100,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X.values)
```

```python
df['k_means_result'] = y_kmeans
```

```python
df.groupby('k_means_result').mean().loc[:,a+['RECENCY','JDT', 'KIDS', 'GD','TR', 'ATV', 'MAX_AMT','TR']].sort_values(['score_predicted'], ascending = False)
```
